## Popularity based recommender system

```python
ratings.merge(book,on="ISBN").shape
```

```python
ratings_with_name=ratings.merge(book,on="ISBN")
```

```python
ratings_with_name
```

```python
#to get how many times a book was rated
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
num_rating_df
```

```python
#to get the average of how many times that book was rated
avg_rating_df=ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
#take only those books which have 250 ratings
popular_df = popular_df[popular_df['num_ratings']>250].sort_values('avg_ratings',ascending=False)
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df.merge(book,on='Book-Title')#some books have same title (see 0&1) and but different isbn because different publication year
#so drop duplicates since rarings is same for all books
```

```python
popular_df = popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-S','Image-URL-M','Image-URL-L','num_ratings','avg_ratings']]
```

## Collaborative Filtering based recommender system

```python
#how many users gave the ratings and how many number of times

ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
#fetch only those users who have give more than 200 ratings
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
knowlegeable_users = x[x].index

```

```python
x[x]
```

```python
knowlegeable_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(knowlegeable_users)]
```

```python
#get only those books which have more than 50 ratings.
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#to find the distance from the book to all the other book including that book itself(treated as a nd plot)
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
similarity_score[0]#gives the similarity scores of the first book and all the userid
```

```python
np.where(pt.index=='Zoya')[0][0]#gives the index of the book
```

```python
#distances = similarity_score[index]find the similarity scores of that book whose index is some x.
sorted(list(enumerate(similarity_score[0])))#get the all the similarity scores of book of index 0 with bookid.
#But sorting will be based on index
#but sorting should be based on the similarity_scores.so,
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1])#get in scending but we want in descending order.
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True)#but first one is the similarity score of the book with itself
sorted(list(enumerate(similarity_score[0])),key=lambda x:x[1],reverse=True)[1:6]#pick top 6 recommendation
```

```python
pt.index
```

```python
def recommend(book_name):
    #index_fetch
    index=np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similar_items:
      print(pt.index[i[0]])
```
